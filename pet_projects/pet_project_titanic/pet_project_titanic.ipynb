```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
titanic_dataset = pd.read_csv("/content/drive/MyDrive/titanic_train.csv")
```

```python
titanic_dataset.head()
```

```python
titanic_dataset.info()
```

```python
titanic_dataset.isnull().sum()
```

```python
titanic_dataset = titanic_dataset.drop(['Cabin'], axis = 1)
```

```python
titanic_dataset.info()
```

```python
titanic_dataset['Age'].fillna(titanic_dataset['Age'].mean(),inplace = True)
```

```python
titanic_dataset.isnull().sum()
```

```python
titanic_dataset['Embarked'].mode()
```

```python
titanic_dataset['Embarked'].fillna(titanic_dataset['Embarked'].mode(), inplace = True)
```

```python
titanic_dataset.isnull().sum()
```

```python
titanic_dataset.describe()
```

```python
titanic_dataset['Survived'].value_counts()
```

```python
titanic_dataset['Embarked'].value_counts()
```

```python
titanic_dataset['Sex'].value_counts()
```

```python
titanic_dataset.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
```

```python
titanic_dataset.head()
```

```python
X = titanic_dataset.drop(['PassengerId','Name','Ticket','Survived'],axis=1)
Y = titanic_dataset['Survived']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state = 13)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
```

```python
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```
