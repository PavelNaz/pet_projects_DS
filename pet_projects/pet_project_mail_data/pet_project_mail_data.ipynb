```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
mail_data = pd.read_csv("/content/drive/MyDrive/mail_data.csv")
```

```python
mail_data.head()
```

```python
mail_data.isnull().sum()
```

```python
mail_data.replace({"Category" :{"ham" : 1, "spam" : 0}}, inplace = True)
```

```python
mail_data.head()
```

```python
X = mail_data['Message']
Y = mail_data['Category']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 13)
```

```python
feature_extraction = TfidfVectorizer(stop_words = "english", lowercase = True)
```

```python
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train_features, Y_train)
```

```python
training_data_prediction = model.predict(X_train_features)
training_accuracy_score = accuracy_score(Y_train, training_data_prediction)
print(training_accuracy_score)
```

```python
test_data_prediction = model.predict(X_test_features)
test_accuracy_score = accuracy_score(Y_test, test_data_prediction)
print(test_accuracy_score)
```

```python
input_mail = ["I love you so much, my dear)"]

# convert text to feature vectors
input_data_features = feature_extraction.transform(input_mail)

# making prediction

prediction = model.predict(input_data_features)
print(prediction)


if (prediction[0]==1):
  print('Ham mail')

else:
  print('Spam mail')
```
