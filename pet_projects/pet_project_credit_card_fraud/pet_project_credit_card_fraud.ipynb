```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('/content/drive/MyDrive/creditcard.csv')
```

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
legit_sample = legit.sample(n=492)
```

```python
new_data = pd.concat([legit_sample,fraud], axis = 0)
```

```python
new_data.head()
```

```python
X = new_data.drop(columns='Class', axis=1)
Y = new_data['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 13)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
training_data_prediction = model.predict(X_train)
training_score = accuracy_score(training_data_prediction, Y_train)
print("training_score : ",training_score)
```

```python
test_data_prediction = model.predict(X_test)
test_score = accuracy_score(test_data_prediction, Y_test)
print("training_score : ",test_score)
```
